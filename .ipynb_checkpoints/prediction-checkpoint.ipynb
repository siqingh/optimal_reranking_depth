```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.patches import Patch
from collections import defaultdict
import json
import numpy as np
import statistics
from sklearn.model_selection import train_test_split, KFold
from utils import convert_run, dcg, dcg_from_run, dcg_for_query, load_umbrela_scores, make_equal_content_bins, assign_to_bin, prettify_label
import matplotlib as mpl
mpl.rcParams["font.family"] = "Times New Roman"
```

```python
data_df = pd.read_csv("data/full_data.tsv", sep="\t")
data_df['max_vs_fully_reranked'] = data_df['max_dcg_value_mono_1000'] / data_df['mono_1000_reranked_all_dcg_value']
optimal_depth_avg = data_df['max_dcg_value_mono_1000_depth'].mean()
max_vs_fully_reranked_ratio_avg = data_df['max_vs_fully_reranked'].mean()
```

```python
def equal_content_prediction_ten_fold_validation(initial_metric_col, fully_reranked_col, retriever, metric, 
                                                 reranker, rerank_depth_array, no_plot=False):
    main_df = data_df[["query_id", initial_metric_col, fully_reranked_col, 
                  f"max_{metric}_value_{reranker}", f"max_{metric}_value_{reranker}_depth"]].copy()

    metric_value_lookup_df = pd.read_csv(f"data/all_queries_all_depth-{retriever}-{reranker}-{metric}.csv")
    metric_value_lookup_df_indexed = metric_value_lookup_df.set_index(["query_id", "depth"])

    # Split all the data into 10 folds
    kf = KFold(n_splits=10, shuffle=True, random_state=42)
    folds = []
    for train_idx, test_idx in kf.split(main_df):
        train_df = main_df.iloc[train_idx]
        test_df = main_df.iloc[test_idx]
        folds.append((train_df, test_df))

    percentiles = [10, 20, 25, 30, 40, 50, 60, 70, 75, 80, 90]
    data = []

    for index, (train_df, test_df) in enumerate(folds):
        train_df = train_df.copy()
        test_df = test_df.copy()

        # Use training data to get equal content bucket
        bin_edges = make_equal_content_bins(train_df[initial_metric_col], n_bins=10)

        # Assign all data to corresponding metric bins
        train_df['metric_bin'] = train_df[initial_metric_col].apply(lambda x: assign_to_bin(x, bin_edges))
        train_df['metric_bin'] = pd.Categorical(
            train_df['metric_bin'],
            ordered=True,
            categories=sorted(set(train_df['metric_bin']), key=lambda x: float(x.split('~')[0]))
        )
        test_df['metric_bin'] = test_df[initial_metric_col].apply(lambda x: assign_to_bin(x, bin_edges))
        test_df['metric_bin'] = pd.Categorical(
            test_df['metric_bin'],
            ordered=True,
            categories=sorted(set(test_df['metric_bin']), key=lambda x: float(x.split('~')[0]))
        )

        # train data's initial metric value metric bin -> rerank depth that maximizes the metric
        train_initial_metric_to_optimal_depths = (
            train_df.groupby("metric_bin", observed=True)[f"max_{metric}_value_{reranker}_depth"]
              .apply(lambda x: sorted(x.tolist()))
              .to_dict()
        )

        for p in percentiles:
            # For each training data metric bin, get all optimal depths
            for metric_bin, depths in train_initial_metric_to_optimal_depths.items():
                # The pth percentile of the optimal depths is the predicted depth for test data in this metric bin
                depth_p = int(np.percentile(depths, p))
        
                # get test queries in this bin
                df_bin = test_df[test_df["metric_bin"] == metric_bin]
        
                for _, row in df_bin.iterrows():
                    if metric == "dcg":
                        metric_at_depth_p = metric_value_lookup_df_indexed.loc[(row['query_id'], depth_p), "metric_value_at_depth"]

                    data.append([index, p, row['query_id'], depth_p, 
                                 row[fully_reranked_col], 
                                 row[initial_metric_col], 
                                 row[f"max_{metric}_value_{reranker}"],
                                 row[f"max_{metric}_value_{reranker}_depth"],
                                 metric_bin,
                                 metric_at_depth_p, 
                                 0 if row[fully_reranked_col] == 0 else metric_at_depth_p / row[fully_reranked_col]])
        
    df = pd.DataFrame(data, columns=['Fold #', 'Percentile', 'Query ID', 'Predicted Depth', 'Reranked All Metric Value', 
                                     'Initial Metric Value', 'Maximum Metric Value', 'Depth of Maximum Metric Value',
                                     'Bin',
                                     'Metric Value At Predicted Depth', 'Ratio (Metric at Depth / Reranked All)'])

    df.to_csv(f'data/equal_content_10_fold-{retriever}-{reranker}-{metric}-{initial_metric_col}.csv', index=False)

    # Rerank all to the same depth
    all_same_depth_avg_predicted_vs_fully_reranked_ratio = {}

    for d in rerank_depth_array:
        # get metrics at depth d
        depth_df = metric_value_lookup_df_indexed.xs(d, level="depth")[
            ["metric_value_at_depth"]
        ].reset_index()
        
        merged = depth_df.merge(main_df, on="query_id", how="inner")
        merged["predicted_vs_fully_reranked"] = (
            merged["metric_value_at_depth"] / merged[fully_reranked_col]
        )
        all_same_depth_avg_predicted_vs_fully_reranked_ratio[d] = merged["predicted_vs_fully_reranked"].mean()

    # Group data by percentile
    grouped_ratio = df.groupby("Percentile").agg(
        avg_predicted_depth=("Predicted Depth", "mean"),
        avg_predicted_vs_fully_reranked_ratio=("Ratio (Metric at Depth / Reranked All)", "mean"),
    ).reset_index()
    grouped_ratio = grouped_ratio.sort_values("avg_predicted_depth")

    if no_plot:
        return all_same_depth_avg_predicted_vs_fully_reranked_ratio, grouped_ratio
        
    if initial_metric_col == "initial_bm25_retrieved_dcg_value":        
        # Plot all same-depth line
        plt.figure(figsize=(9, 6))
        plt.plot(list(all_same_depth_avg_predicted_vs_fully_reranked_ratio.keys()),
                 list(all_same_depth_avg_predicted_vs_fully_reranked_ratio.values()), 
                 marker='o', linestyle='-', linewidth=2, label="All To Same Depth")
        plt.scatter(optimal_depth_avg, max_vs_fully_reranked_ratio_avg, color='red', marker='o', label='Optimal')
        plt.xlabel(f"Avg Reranking Depth (Mono-T5)", fontsize=13, labelpad=8)
        plt.ylabel(f"Avg DCG@10 Ratio (@ Rerank Depth / @ Fully Reranked)", fontsize=13, labelpad=8)
        plt.grid(True, which="both", ls="--", alpha=0.5)
        plt.legend()
        plt.savefig(f"plot/fixed_depth_reranking.png", dpi=300)
        plt.show()
        plt.close()
    
    # Plot average ratio (denominator fully reranked)
    plt.figure(figsize=(9, 6))
    # Plot all same-depth line
    plt.plot(list(all_same_depth_avg_predicted_vs_fully_reranked_ratio.keys()),
             list(all_same_depth_avg_predicted_vs_fully_reranked_ratio.values()), 
             marker='o', linestyle='-', linewidth=2, label="All To Same Depth")
    # Plot percentile prediction points
    for _, row in grouped_ratio.iterrows():
        p = row["Percentile"]
        x = row["avg_predicted_depth"]
        y = row["avg_predicted_vs_fully_reranked_ratio"]  # or whichever ratio col you want
        plt.scatter(x, y, label=f"thr={p/100:.2f}", s=80, color=cmap(percentiles.index(p)))
    
    plt.xlabel(f"Reranking Depth ({prettify_label(reranker)})", fontsize=13, labelpad=8)
    plt.ylabel(f"Avg {prettify_label(metric)} Ratio (@ Predicted Depth / @ Fully Reranked)", fontsize=13, labelpad=8)
    plt.title(f"Predicted Using {prettify_label(initial_metric_col)}", fontsize=14, pad=12)
    plt.grid(True, which="both", ls="--", alpha=0.5)
    plt.legend()
    plt.savefig(f"plot/equal_content_10_fold-avg_predicted_vs_fully_reranked_ratio-{metric}-{retriever}-{reranker}-{initial_metric_col}.png", dpi=300)
    #plt.show()
    plt.close()

    return all_same_depth_avg_predicted_vs_fully_reranked_ratio, grouped_ratio
```

```python
_, _ = equal_content_prediction_ten_fold_validation("initial_bm25_retrieved_dcg_value", "mono_1000_reranked_all_dcg_value",
                                             "bm25", "dcg", "mono_1000", [0, 1, 2, 5, 10, 20, 30, 50, 100, 250, 500, 800, 1000])
```

```python
initial_metric_col = "initial_bm25_retrieved_dcg_value"
all_same_depth_avg_predicted_vs_fully_reranked_ratio, grouped_ratio = equal_content_prediction_ten_fold_validation(
    initial_metric_col, "mono_1000_reranked_all_dcg_value", "bm25", "dcg", "mono_1000", [0, 1, 2, 5, 10, 20, 30, 50, 100, 250, 500, 800, 1000],
    True)

fig, axes = plt.subplots(1, 2, figsize=(9, 4.5))

# Plot 1
ax1 = axes[0]
ax1.plot(list(all_same_depth_avg_predicted_vs_fully_reranked_ratio.keys()),
         list(all_same_depth_avg_predicted_vs_fully_reranked_ratio.values()), 
         linewidth=1.5, marker='o', markersize=4, label="All To Same Depth", zorder=1, alpha=0.6)
for _, row in grouped_ratio.iterrows():
    p = row["Percentile"]
    x = row["avg_predicted_depth"]
    y = row["avg_predicted_vs_fully_reranked_ratio"]
    ax1.scatter(x, y, s=30, color=cmap(percentiles.index(p)), label=f"thr={p/100:.2f}", zorder=2)

ax1.set_ylabel(f"Avg DCG@10 Ratio (@ Predicted Depth / @ Fully Reranked)")
ax1.grid(True, which="both", ls="--", alpha=0.5)

# Zoomed in Plot
ax2 = axes[1]
orig_x_vals = [k for k in all_same_depth_avg_predicted_vs_fully_reranked_ratio.keys() if k >= 100]
y_vals = [all_same_depth_avg_predicted_vs_fully_reranked_ratio[k] for k in orig_x_vals]
ax2.plot(orig_x_vals, y_vals, marker='o', linestyle='-', linewidth=1.5, markersize=4, alpha=0.6, label="All To Same Depth")
for _, row in grouped_ratio.iterrows():
    p = row["Percentile"]
    orig_x = row["avg_predicted_depth"]
    if orig_x >= 100:
        y = row["avg_predicted_vs_fully_reranked_ratio"]
        ax2.scatter(orig_x, y, s=30, color=cmap(percentiles.index(p)), label=f"thr={p/100:.2f}")
ax2.scatter(optimal_depth_avg, max_vs_fully_reranked_ratio_avg, color='red', marker='o', label='Optimal')

# Bottom axis = depth
ax2.grid(True, which="both", ls="--", alpha=0.5)
ax2.set_xlim(100, max(orig_x_vals))

# Top axis = reduction factor
secax = ax2.secondary_xaxis(
    'top',
    functions=(lambda x: 1000/x, lambda x: 1000/x)
)
secax.set_xlabel("Reduction Factor (1000 / Reranking Depth)")

tick_depths = np.arange(100, max(orig_x_vals)+1, 200)
ax2.set_xticks(tick_depths)  # bottom axis = depths

#fig.suptitle(f"Predicted Using {prettify_label(initial_metric_col)}")

handles = []
labels = []
for ax in axes:
    h, l = ax.get_legend_handles_labels()
    handles.extend(h)
    labels.extend(l)
unique_labels = []
unique_handles = []
for handle, label in zip(handles, labels):
    if label not in unique_labels:
        unique_labels.append(label)
        unique_handles.append(handle)
    
ax2.legend(handles=unique_handles, labels=unique_labels, loc='upper right', ncol=2)

fig.text(0.5, 0.01, "Avg Reranking Depth (Mono-T5)", ha='center', va='center')

plt.tight_layout()
plt.savefig(f"plot/{initial_metric_col}_performance.png", dpi=300)
plt.show()
plt.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle, ConnectionPatch

fig, ax = plt.subplots(figsize=(9, 4.5))

# Main plot, full axis
x_all = list(all_same_depth_avg_predicted_vs_fully_reranked_ratio.keys())
y_all = list(all_same_depth_avg_predicted_vs_fully_reranked_ratio.values())

ax.plot(x_all, y_all,
        linewidth=1.5, marker='o', markersize=4,
        label="All To Same Depth", zorder=1, alpha=0.6)
ax.scatter(optimal_depth_avg, max_vs_fully_reranked_ratio_avg, color='red', marker='o', label='Optimal')

for _, row in grouped_ratio.iterrows():
    p = row["Percentile"]
    x = row["avg_predicted_depth"]
    y = row["avg_predicted_vs_fully_reranked_ratio"]
    ax.scatter(x, y, s=30, color=cmap(percentiles.index(p)),
               label=f"thr={p/100:.2f}", zorder=2)

ax.set_ylabel("Avg DCG@10 Ratio (@ Predicted Depth / @ Fully Reranked)")
ax.grid(True, which="both", ls="--", alpha=0.5)

# Determine Zoom Region
zoom_xmin = 100
zoom_xmax = 510

orig_x_vals = sorted(list(all_same_depth_avg_predicted_vs_fully_reranked_ratio.keys()))
zoom_mask = [x for x in orig_x_vals if zoom_xmin <= x <= zoom_xmax]
zoom_y_vals = [all_same_depth_avg_predicted_vs_fully_reranked_ratio[k] for k in zoom_mask]

zoom_ymin = min(zoom_y_vals) * 0.98
zoom_ymax = max(zoom_y_vals) * 1.02

# Draw dotted rectangle on main plot
rect = Rectangle((zoom_xmin, zoom_ymin),
                 width=(zoom_xmax - zoom_xmin),
                 height=(zoom_ymax - zoom_ymin),
                 facecolor='none',
                 edgecolor='gray',
                 linestyle=':',
                 linewidth=1.5,
                 zorder=3,
                 alpha=0.9)
ax.add_patch(rect)

# Create zoomed plot
# [left, bottom, width, height] in figure fraction (0..1).
axins = fig.add_axes([0.62, 0.2, 0.34, 0.34])  

axins.plot([k for k in orig_x_vals if k >= zoom_xmin],
           [all_same_depth_avg_predicted_vs_fully_reranked_ratio[k] for k in orig_x_vals if k >= zoom_xmin],
           marker='o', linestyle='-', linewidth=1.5, markersize=4, alpha=0.6, label="All To Same Depth")

for _, row in grouped_ratio.iterrows():
    p = row["Percentile"]
    orig_x = row["avg_predicted_depth"]
    if zoom_xmin <= orig_x <= zoom_xmax:
        y = row["avg_predicted_vs_fully_reranked_ratio"]
        axins.scatter(orig_x, y, s=30, color=cmap(percentiles.index(p)), label=f"thr={p/100:.2f}")

axins.set_xlim(zoom_xmin, zoom_xmax)
axins.set_ylim(zoom_ymin, zoom_ymax)

# Set ticks (for reduction factor 9 ~ 2)
rf_labels = np.arange(9, 1, -1)  
depth_ticks = [1000 / rf for rf in rf_labels]
axins.set_xticks(depth_ticks)

axins.set_xticklabels(rf_labels)
axins.set_xlabel("Reduction Factor (1000 / Reranking Depth)")

axins.grid(True, which="major", axis="x", ls="--", alpha=0.5)
axins.grid(True, which="major", axis="y", ls="--", alpha=0.25)

show_labels = [1000/x for x in [9, 6, 5, 4, 3, 2]]
depth_labels = [str(int(d)) if d in show_labels else "" for d in depth_ticks]

secax = axins.secondary_xaxis('top', functions=(lambda x: x, lambda x: x))
secax.set_xticks(depth_ticks)
secax.set_xticklabels(depth_labels)
secax.set_xlabel("Avg Reranking Depth")

for tl in secax.get_xticklabels():
    tl.set_fontsize('small')
for tl in axins.get_xticklabels():
    tl.set_fontsize('small')


# Draw dotted connector lines from the rectangle to the zoomed plot
cp1 = ConnectionPatch(xyA=(zoom_xmax, zoom_ymax), coordsA=ax.transData,
                      xyB=(0, 1), coordsB=axins.transAxes,
                      linestyle=':', linewidth=1.25, color='gray')
cp2 = ConnectionPatch(xyA=(zoom_xmax, zoom_ymin), coordsA=ax.transData,
                      xyB=(0, 0), coordsB=axins.transAxes,
                      linestyle=':', linewidth=1.25, color='gray')

fig.add_artist(cp1)
fig.add_artist(cp2)

ax.legend(loc='upper right', ncol=4, fontsize='small')

# Shared x-label and final layout/save
fig.text(0.5, 0.02, "Avg Reranking Depth (Mono-T5)", ha='center', va='center')
plt.tight_layout()
plt.savefig(f"plot/{initial_metric_col}_performance.png", dpi=300)
plt.show()
plt.close()
```
